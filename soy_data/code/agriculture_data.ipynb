```python
import pandas as pd
import glob, os 
```

```python
path = r'/Users/dianabaptistarojo/Documents/Columbia/masters_project/siap_data'                   
all_files = glob.glob(os.path.join(path, "*.csv"))

df_from_each_file = (pd.read_csv(f, encoding='latin-1') for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
concatenated_df.tail(10)
```

```python
concatenated_df.to_csv('total_data_siap.csv', index=False)
```

```python
concatenated_df.loc[concatenated_df['Anio'] == 2017]
```

```python
soy_nac = concatenated_df.loc[concatenated_df['Nomcultivo'] == 'Soya']
```

¿Cómo evolucionó la siembra de soya a nivel nacional, por estado? // 
What's the trend of planted soy at a national level?

```python
soy_year_state = soy_nac.groupby(['Nomestado', 'Anio']).sum()['Sembrada']
```

```python
soy_year_state
```

# Planted soy by year in Mexico

```python
year_soy = soy_nac.groupby('Anio')['Sembrada'].sum().reset_index()
```

```python
year_soy
```

```python
year_soy.to_csv('soy_production_by_year.csv', index=False)
```

```python
%matplotlib inline
```

```python
year_soy.plot(x='Anio', y='Sembrada', kind='line')
```
